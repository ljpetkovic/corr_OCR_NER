### Répartition des vrais/faux positifs/négatifs et les tokens incorrectement classifiés comme les EN



```python
import pandas as pd

import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv("boxplot.csv")

# display 5 rows of dataset
# df.head()

del df["id"]
ax = df.plot(ylabel ="Nombre d'EN",xlabel="Type d'EN",kind="bar")
ax.set_xticks(df.index)
ax.set_xticklabels(df.type_EN,rotation=0)
ax.legend(["Nombre d'EN"])

```
